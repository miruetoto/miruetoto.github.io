```python
%%R 
rdata <- pd %>% 
        group_by(sourceid,dstid) %>% 
        summarize(mean_travel=mean(mean_travel_time)) %>% 
        mutate(W=exp(-mean_travel/1000))
```

```python
%%R
W<-rep(0,608*608); dim(W)<-c(608,608)
for (i in 1:608){
    for(j in 1:608){
        if(dim(rdata %>% filter(sourceid==i,dstid==j))[1]==0) W[i,j]<-0
        else W[i,j] <- (rdata %>% filter(sourceid==i,dstid==j))$W
    }
    print(i)
}
```
